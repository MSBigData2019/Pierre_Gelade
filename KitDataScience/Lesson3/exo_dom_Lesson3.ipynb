```python
# coding: utf-8
import requests,re
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def ExtractListActive():
    listNoms = []
    url = r"https://gist.github.com/paulmillr/2657075"
    res = requests.get(url)
    if res.status_code == 200:
        html_doc =  res.text
        soup = BeautifulSoup(html_doc,"html.parser")
        t = soup.find("table").findAll("tr")[1:]
        return [l.find("td").text.split()[0] for l in t]
    
def ExtractJson2(user,listStars,nbpagesmax):
    token = ""
    headers = {'Authorization': 'token {}'.format(token)}
    for p in range(nbpagesmax):
        url = f"https://api.github.com/users/{user}/repos?page={p}&per_page=100"
        req = requests.get(url,headers=headers)
        page = req.json()
        ret = [(repos["stargazers_count"]) for repos in page]
        for s in ret:
            listStars.append(s)
    return listStars
```

```python
DicoUsersStars = {}
nomsUser = ExtractListActive()
#nomsUser = ['fabpot','andrew','taylorotwell','egoist','HugoGiraudel']
#nomsUser = ['fabpot','andrew','taylorotwell']
for user in nomsUser:
    listStars = []
    nbpagesmax = 10
    Stars = ExtractJson2(user,listStars,nbpagesmax)
    DicoUsersStars[user] = round(sum(Stars)/len(Stars),2)
    print(user,round(sum(Stars)/len(Stars),2),len(Stars))

sorted(DicoUsersStars.items(), key=lambda t: t[1])
```
